Categories are imported and WikiProjects are identified through them.

```python
df_ca = pd.read_csv('data/category.tsv', sep='\t')
df_ca
```

```python
df_fa = df_ca[df_ca['title'].str.contains('^FA-Class_.*_articles$', regex= True, na=False)]
df_fa.loc[:,'class'] = 'FA'
df_fl = df_ca[df_ca['title'].str.contains('^FL-Class_.*_articles$', regex= True, na=False)]
df_fl.loc[:,'class'] = 'FL'
df_ga = df_ca[df_ca['title'].str.contains('^GA-Class_.*_articles$', regex= True, na=False)]
df_ga.loc[:,'class'] = 'GA'
df_aa = df_ca[df_ca['title'].str.contains('^A-Class_.*_articles$', regex= True, na=False)]
df_aa.loc[:,'class'] = 'A'
df_ba = df_ca[df_ca['title'].str.contains('^B-Class_.*_articles$', regex= True, na=False)]
df_ba.loc[:,'class'] = 'B'
df_cca = df_ca[df_ca['title'].str.contains('^C-Class_.*_articles$', regex= True, na=False)]
df_cca.loc[:,'class'] = 'C'
df_sa = df_ca[df_ca['title'].str.contains('^Start-Class_.*_articles$', regex= True, na=False)]
df_sa.loc[:,'class'] = 'Start'
df_sta = df_ca[df_ca['title'].str.contains('^Stub-Class_.*_articles$', regex= True, na=False)]
df_sta.loc[:,'class'] = 'Stub'
df_la = df_ca[df_ca['title'].str.contains('^List-Class_.*_articles$', regex= True, na=False)]
df_la.loc[:,'class'] = 'List'
df_as = pd.concat([df_fa, df_fl, df_ga, df_aa, df_ba, df_cca, df_sa, df_sta, df_la])
```

```python
df_cl = pd.read_csv('data/page_category.tsv', sep='\t')
df_cl
```

## Network

```python
total_wp = df_pages_m.drop(columns=['title_x', 'class']).drop_duplicates()
total_wp['title_y'] = total_wp['title_y'].replace(to_replace='^(FA|GA|A|B|C|Start|Stub|FL|List)-Class_', value='', regex=True)
total_wp['title_y'] = total_wp['title_y'].replace(to_replace='_articles$', value='', regex=True)
total_wp['title_y'] = total_wp['title_y'].replace(to_replace='_', value=' ', regex=True)
total_wp = total_wp.drop_duplicates()
total_wp = total_wp[~total_wp['title_y'].isin(['Featured topics',
                                               'Version 0.7',
                                               'Version 1.0',
                                               'Version 1.0',
                                               'Version 1.0 vital'])]
len(set(total_wp.page_id.tolist()))
```

```python
len(set(total_wp.title_y.tolist()))
```

To avoid noise, we have selected only pages classified as FA, GA, A, B, and C. A total of 525,390 pages (8.30%).

```python
df_pages_m = df_pages_m[df_pages_m['class'].isin(['FA','GA','A','B','C'])]
df_pages_m = df_pages_m.drop(columns=['title_x', 'class']).drop_duplicates()
len(set(df_pages_m.page_id.tolist()))
```

```python
df_pages_m['title_y'] = df_pages_m['title_y'].replace(to_replace='^(FA|GA|A|B|C)-Class_', value='', regex=True)
df_pages_m['title_y'] = df_pages_m['title_y'].replace(to_replace='_articles$', value='', regex=True)
df_pages_m['title_y'] = df_pages_m['title_y'].replace(to_replace='_', value=' ', regex=True)
df_pages_m = df_pages_m.drop_duplicates()
df_pages_m
```

```python
df_pages_m = df_pages_m[~df_pages_m['title_y'].isin(['Featured topics',
                                               'Version 0.7',
                                               'Version 1.0',
                                               'Version 1.0',
                                               'Version 1.0 vital'])]
df_pages_m
```

```python
df_pages_m.title_y = df_pages_m.title_y.str.lower()
```

```python
df_l = pd.read_csv('data/page_link.tsv', sep='\t')
df_l
```

```python
df_l = df_l[(df_l['page_id'].isin(df_pages_m['page_id'].tolist())) &
           (df_l['linked_page_id'].isin(df_pages_m['page_id'].tolist()))]
df_l
```

```python
df_l['source'] = df_l[['page_id', 'linked_page_id']].min(axis=1)
df_l['target'] = df_l[['page_id', 'linked_page_id']].max(axis=1)
df_l = df_l[['source', 'target']].drop_duplicates()
df_l
```

```python
df_l = df_l.merge(df_pages_m[['page_id', 'title_y']].drop_duplicates(), how='inner', left_on='source', right_on='page_id')[['title_y', 'target']]
df_l = df_l.merge(df_pages_m[['page_id', 'title_y']].drop_duplicates(), how='inner', left_on='target', right_on='page_id')[['title_y_x', 'title_y_y']]
df_l
```

```python
df_l = df_l[df_l['title_y_x'] != df_l['title_y_y']]
df_l['source'] = df_l[['title_y_x', 'title_y_y']].min(axis=1)
df_l['target'] = df_l[['title_y_x', 'title_y_y']].max(axis=1)
df_l = df_l[['source', 'target']]
df_l
```

```python
df_l = df_l.groupby(['source', 'target']).size().reset_index(name='weight')
df_l
```

```python
df_l.to_csv('results/wikiprojects_network.txt', sep='\t', index=False, header=False)
```

```python
df_pages_m = df_pages_m[['title_y', 'doi', 'isbn', 'references']]
df_pages_m['pages'] = 1
df_pages_m['pages_doi'] = 0
df_pages_m.loc[df_pages_m['doi']>0,'pages_doi'] = 1
```

```python

df_pages_m['pages_isbn'] = 0
df_pages_m.loc[df_pages_m['isbn']>0,'pages_isbn'] = 1
df_pages_m = df_pages_m.groupby('title_y').agg({'pages':'sum',
                                                'doi':'mean',
                                                'pages_doi':'sum',
                                                'isbn':'mean',
                                                'pages_isbn':'sum',
                                                'references':'mean'
                              }).reset_index()
df_pages_m
```

```python
df_pages_m.pages_doi = round(100*df_pages_m.pages_doi/df_pages_m.pages, 3)
df_pages_m.pages_isbn = round(100*df_pages_m.pages_isbn/df_pages_m.pages, 3)
df_pages_m
```

```python
df_pages_m.rename(columns={'title_y':'id',
                           'pages':'weight<pages>',
                           'doi':'score<doi>',
                           'pages_doi':'score<pages doi>',
                           'isbn':'score<isbn>',
                           'pages_isbn':'score<pages isbn>',
                           'references':'score<references>'
                          },
          inplace=True)
```

```python
df_pages_m.to_csv('results/wikiprojects_map.txt', sep='\t', index=False)
```
