```python
import pandas as pd
import csv
```

```python
df_page = pd.read_csv('data/page.tsv', sep='\t', quoting=csv.QUOTE_NONE)
df_page = df_page[(df_page['namespace']==0) & (df_page['is_redirect']==0)]
df_page
```

```python
df_pubs = pd.read_csv('data/pub.tsv', sep='\t')
df_pubs
```

```python
df_page_pubs = pd.read_csv('data/page_pub.tsv', sep='\t')
df_page_pubs
```

```python
doi = df_page_pubs[df_page_pubs['pub_id'].isin(df_pubs[(~df_pubs['doi'].isna()) & (df_pubs['isbn'].isna())].pub_id.tolist())].groupby('page_id').size().reset_index(name='doi')
isbn = df_page_pubs[df_page_pubs['pub_id'].isin(df_pubs[~df_pubs['isbn'].isna()].pub_id.tolist())].groupby('page_id').size().reset_index(name='isbn')
refs = df_page_pubs[df_page_pubs['pub_id'].isin(df_pubs[(~df_pubs['isbn'].isna()) | (~df_pubs['doi'].isna())].pub_id.tolist())].groupby('page_id').size().reset_index(name='references')
df_page = df_page[['page_id','title']].merge(doi, how='left', on='page_id').merge(isbn, how='left', on='page_id').merge(refs, how='left', on='page_id')
df_page
```

```python
df_page.doi = df_page.doi.fillna(0)
df_page.isbn = df_page.isbn.fillna(0)
df_page.references = df_page.references.fillna(0)
df_page[['page_id', 'doi', 'isbn', 'references']] = df_page[['page_id', 'doi', 'isbn', 'references']].astype(int)
df_page
```

```python
df_page.to_csv('results/page_references.tsv', index=False, sep='\t')
```
